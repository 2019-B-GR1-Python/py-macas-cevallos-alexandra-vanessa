```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
from scipy import misc
```

```python
mapache = misc.face()
mapache
```

```python
plt.imshow(mapache)
plt.show
```

```python
#todo rojo
mapache_red=np.copy(mapache)
mapache_red[:,:,1]=0
mapache_red[:,:,2]=0
plt.title("mapache canal rojo")
plt.imshow(mapache_red)
```

```python
#todo verde
mapache_green=np.copy(mapache) 
mapache_green[:,:,2]=0
plt.title("mapache_green_ sin canal azul")
plt.imshow(mapache_green)
```

```python
#Para hacer una inspeccin detallada de variaciones de intensidad podemos usar interpolation='nearest'
mapache_inspeccion=np.copy(mapache)
plt.imshow(mapache_inspeccion[100:800, 250:600], cmap=plt.cm.gray)
plt.imshow(mapache_inspeccion[100:800, 250:600], cmap=plt.cm.gray, interpolation='nearest')
```
