```python
import os
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
```

```python
mapache = misc.face()
tamanio_tablero_n = input("Ingrese el tama√±o del tablero")
mapache_parts = np.vsplit(mapache,int(tamanio_tablero_n))
#print(mapache_parts)
piezas_index = np.random.permutation(int(tamanio_tablero_n))
mapache_random = np.concatenate(list(map(lambda x: mapache_parts[x],piezas_index)))

#print(mapache_random)
print("Origin")
plt.imshow(mapache)
plt.show()
plt.imshow(mapache_random)
plt.show()

def menu():
    os.system('clear')
    print ("Selecciona una opci√≥n")
    print ("\t1 - Jugar")
    print ("\t2 - Salir")

while True:
    menu()
    opcionMenu = input("Ingresa una opcion: ")
    
    if opcionMenu=="1":
        print("Play")
        mover_posicion_a = input("Ficha a mover: ")
        mover_posicion_b = input("Posicion a mover: ")
        pieza_temporal = piezas_index[int(mover_posicion_a)]
        piezas_index[int(mover_posicion_a)] = piezas_index[int(mover_posicion_b)]
        piezas_index[int(mover_posicion_b)] = pieza_temporal
        
        mapache_desordenado = np.concatenate(list(map(lambda x: mapache_parts[x],piezas_index)))
        plt.imshow(mapache_desordenado)
        plt.show()
        
    
    elif opcionMenu=="2":
        print("Salio del rompecabezas")
        break
    

 
        
```
