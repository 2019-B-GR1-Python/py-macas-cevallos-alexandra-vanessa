1) Crea un Dataframe de 10 registros y 6 columnas y consigue las 5 primeros y los 5 ultimos registros

```python
import pandas as pd
import os
import numpy as np
df1 = pd.DataFrame(np.random.randint(1, 10, 60).reshape(-1, 6), columns=list('ABCDEF'))
df1
```

```python
def func(n, condicion): 
    if(condicion == True):
        return(df1[:n])
    else:
        return(df1[n+1:])
print('primeros')
print(func(5,True)) # los primeros 5
print('ultimos')
print(func(5,False)) # los 5 ultimos


```

7) Crear un Dataframe con 10 registros y 6 columnas y Ordenar el dataframe 1 vez por cada columna, ascendente y descendente

```python
df7 = pd.DataFrame(np.random.randint(1, 10, 60).reshape(-1, 6), columns=list('ABCDEF'))
df7
```

```python
def func7(condicion,df,col): 
    return(df.sort_values(col,ascending=condicion))

func7(False,df7,'A') # TRUE ACENDENTE Y FALSE DESENDENTE el a se cambia por la columna
```

```python
func7(True,df7,'B') # TRUE ACENDENTE Y FALSE DESENDENTE el a se cambia por la columna
```
